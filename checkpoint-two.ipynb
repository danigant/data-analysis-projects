## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#Taking a look at the distribution of IMDb Scores and TMDb Scores in the titles data frame:
fig, axes = plt.subplots(1,2, figsize=(15,6))
#Histogram 1 IMDb Scores
axes[0].hist(titles_df["imdb_score"], bins=20, color="skyblue", edgecolor= "black")
axes[0].set_title("Distribution of IMDb Scores")

#Histogram 2 TMDb Scores
axes[1].hist(titles_df["tmdb_score"], bins=20, color="orange", edgecolor= "black")
axes[1].set_title("Distribution of TMDb Scores")

plt.ylabel("Frequency")
plt.show()
```

```python
#Look at the coordination of IMDb Scores and IMDb Votes:
plt.figure(figsize=(10,6))
plt.scatter(titles_df["imdb_score"], titles_df["imdb_votes"], color='dodgerblue', alpha=0.5)
plt.title("IMDb Score vs IMDb Votes")
plt.xlim(0, 10)
plt.xlabel("IMDb Score")
plt.ylabel("IMDb Votes (in millions)")
plt.show()
```

```python
#Looking at IMDb score over time:
plt.figure(figsize=(10,6))
plt.scatter(titles_df['release_year'], titles_df['imdb_score'], color='dodgerblue', alpha=0.5)
plt.title("IMDb Score Over Time")
plt.xlabel("Release Year")
plt.ylabel("IMDb Score")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(titles_df['release_year'], titles_df['imdb_votes'], color='dodgerblue', alpha=0.5)
plt.title("IMDb Votes Over Time")
plt.xlabel("Release Year")
plt.ylabel("IMDb Votes (in millions)")
plt.show()
```
